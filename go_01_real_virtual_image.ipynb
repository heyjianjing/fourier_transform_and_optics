### Real and virtual image

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Arc, Rectangle, Ellipse, Circle
import numpy as np

plt.style.use('dark_background')
# color: https://matplotlib.org/stable/gallery/color/named_colors.html
```

```python
def convex_lens_imaging_real(ax, title=None, title_loc='center', fontsize=12, annotation=True):
    # Optical axis
    ax.plot([-1.2, 1.2], [0, 0], 'w', linewidth=2)

    # Distances
    d = 0.05
    f = 0.4

    # y1 and arrows
    y1 = 0.4
    ax.plot([-2.5*f, -2.5*f], [0, y1], 'c', linewidth=2)
    ax.arrow(-2.5*f, y1-0.1, 0, 0,  head_width=0.07, head_length=0.1, fc='c', ec='none')

    # Angles
    theta_1 = 0
    theta_2 = np.arctan2(y1, f-d/2)

    # Ray 1
    ax.plot([-2.5*f, -d/2], [y1, y1], 'r', linewidth=2)
    ax.plot([d/2, 2.5*f], [y1, y1-(2.5*f-d/2)*np.tan(theta_2)], 'r', linewidth=2)
    ax.arrow(-f, y1, 0.001, 0,  head_width=0.04, head_length=0.06, fc='w', ec='none', zorder=5)
    ax.arrow(d/2+f/2, y1-f/2*np.tan(theta_2), 0.001*np.cos(theta_2), -0.001*np.sin(theta_2),  head_width=0.04, head_length=0.06, fc='w', ec='none', zorder=5)

    # Ray1 2
    ax.plot([-2.5*f, 2.5*f], [y1, -y1], 'r', linewidth=2)
    ax.arrow(-f, y1/2.5, 0.0025, -0.001,  head_width=0.04, head_length=0.06, fc='w', ec='none', zorder=5)

    # Ray 3
    ax.plot([-2.5*f, -d/2], [y1, y1-(2.5*f-d/2)/1.5], 'r', linewidth=2)
    ax.plot([d/2, 2.5*f], [y1-(2.5*f-d/2)/1.5, y1-(2.5*f-d/2)/1.5], 'r', linewidth=2)
    ax.arrow(0.5*f, y1-(2.5*f-d/2)/1.5, 0.001, 0,  head_width=0.04, head_length=0.06, fc='w', ec='none', zorder=5)

    # y2 and arrows
    y2 = 0.4
    ax.plot([(f-d/2)/y1*(2.5*f-d/2)/1.5+d/2, (f-d/2)/y1*(2.5*f-d/2)/1.5+d/2], [0, y1-(2.5*f-d/2)/1.5], 'lime', linewidth=2, zorder=5)
    ax.arrow((f-d/2)/y1*(2.5*f-d/2)/1.5+d/2+0.005, y1-(2.5*f-d/2)/1.5+0.08, 0, -0.001,  head_width=0.07, head_length=0.1, fc='lime', ec='none')

    # f distance
    ax.plot([0, 0], [0, -0.1], 'w-', linewidth=1)
    ax.plot([0, f], [-0.05, -0.05], 'w-', linewidth=1)
    ax.plot([f, f], [0, -0.1], 'w-', linewidth=1)
    ax.plot([0, -f], [-0.05, -0.05], 'w-', linewidth=1)
    ax.plot([-f, -f], [0, -0.1], 'w-', linewidth=1)
    ax.text(f/2, -0.15, r'$f$', horizontalalignment='center', verticalalignment='center', fontsize=fontsize, color='w')
    ax.text(-f/2, -0.15, r'$f$', horizontalalignment='center', verticalalignment='center', fontsize=fontsize, color='w')
    ax.arrow(0.06, -0.05, -0.001, 0,  head_width=0.04, head_length=0.06, fc='w', ec='none')
    ax.arrow(f-0.06, -0.05, 0.001, 0,  head_width=0.04, head_length=0.06, fc='w', ec='none')
    ax.arrow(-0.06, -0.05, 0.001, 0,  head_width=0.04, head_length=0.06, fc='w', ec='none')
    ax.arrow(-f+0.06, -0.05, -0.001, 0,  head_width=0.04, head_length=0.06, fc='w', ec='none')

    if annotation:
        # s distance
        ax.plot([0, 0], [y1-(2.5*f-d/2)/1.5, y1-(2.5*f-d/2)/1.5-0.1], 'w-', linewidth=1)
        ax.plot([0, (f-d/2)/y1*(2.5*f-d/2)/1.5+d/2], [y1-(2.5*f-d/2)/1.5-0.05, y1-(2.5*f-d/2)/1.5-0.05], 'w-', linewidth=1)
        ax.plot([(f-d/2)/y1*(2.5*f-d/2)/1.5+d/2, (f-d/2)/y1*(2.5*f-d/2)/1.5+d/2], [y1-(2.5*f-d/2)/1.5, y1-(2.5*f-d/2)/1.5-0.1], 'w-', linewidth=1)
        ax.plot([0, -2.5*f], [y1-(2.5*f-d/2)/1.5-0.05, y1-(2.5*f-d/2)/1.5-0.05], 'w-', linewidth=1)
        ax.plot([-2.5*f, -2.5*f], [y1-(2.5*f-d/2)/1.5, y1-(2.5*f-d/2)/1.5-0.1], 'w-', linewidth=1)
        ax.text(((f-d/2)/y1*(2.5*f-d/2)/1.5+d/2)/2, y1-(2.5*f-d/2)/1.5-0.15, r'$s_i$', horizontalalignment='center', verticalalignment='center', fontsize=fontsize, color='w')
        ax.text(-2.5*f/2, y1-(2.5*f-d/2)/1.5-0.15, r'$s_o$', horizontalalignment='center', verticalalignment='center', fontsize=fontsize, color='w')
        ax.arrow(0.06, y1-(2.5*f-d/2)/1.5-0.05, -0.001, 0,  head_width=0.04, head_length=0.06, fc='w', ec='none')
        ax.arrow((f-d/2)/y1*(2.5*f-d/2)/1.5+d/2-0.06, y1-(2.5*f-d/2)/1.5-0.05, 0.001, 0,  head_width=0.04, head_length=0.06, fc='w', ec='none')
        ax.arrow(-0.06, y1-(2.5*f-d/2)/1.5-0.05, 0.001, 0,  head_width=0.04, head_length=0.06, fc='w', ec='none')
        ax.arrow(-2.5*f+0.06, y1-(2.5*f-d/2)/1.5-0.05, -0.001, 0,  head_width=0.04, head_length=0.06, fc='w', ec='none')

    # Lens
    lens = Ellipse((0,0), d, 1.2, edgecolor='none', facecolor='skyblue', zorder=-5)
    ax.add_patch(lens)

    # Set limits and aspect ratio
    ax.set_xlim(-1.2, 1.2)
    ax.set_ylim(-0.7, 0.7)
    ax.set_aspect('equal')
    # ax.grid(True)
    ax.axis('off')

    if title:
        ax.set_title(title, loc=title_loc, fontsize=fontsize)

def convex_lens_imaging_virtual(ax, title=None, title_loc='center', fontsize=12, annotation=True):
    # Optical axis
    ax.plot([-1.2, 1.2], [0, 0], 'w', linewidth=2)

    # Distances
    d = 0.05
    f = 0.4

    # y2 and arrows
    y2 = 0.4
    ax.plot([-2.5*f, -2.5*f], [0, y2], 'lime', linewidth=2)
    ax.arrow(-2.5*f, y2-0.1, 0, 0,  head_width=0.07, head_length=0.1, fc='lime', ec='none')

    # # Angles
    # theta_1 = 0
    # theta_2 = np.arctan2(y2, f-d/2)

    # Ray 1
    ax.plot([-2.5*f, -d/2], [y2, y2], 'r--', linewidth=2)
    ax.plot([d/2, 2.5*f], [y2, y2], 'r', linewidth=2)
    ax.arrow(f, y2, 0.001, 0,  head_width=0.04, head_length=0.06, fc='w', ec='none', zorder=5)

    # Ray 2
    ax.plot([-2.5*f, 0], [y2, y2-y2/(3.5*f)*2.5*f], 'r--', linewidth=2)
    ax.plot([0, 2.5*f], [y2-y2/(3.5*f)*2.5*f, y2-2*y2/(3.5*f)*2.5*f], 'r', linewidth=2)
    ax.plot([-2.5*f+2.5*f/(2*y2-y2/(3.5*f)*2.5*f)*y2-0.05, 0], [y2-y2/(3.5*f)*2.5*f, y2-y2/(3.5*f)*2.5*f], 'r', linewidth=2)
    ax.arrow(f, 0, 0.001*2.5*f, -0.001*y2/(3.5*f)*2.5*f,  head_width=0.04, head_length=0.06, fc='w', ec='none', zorder=5)

    # y1 and arrows
    ax.plot([-2.5*f+2.5*f/(2*y2-y2/(3.5*f)*2.5*f)*y2-0.05, -2.5*f+2.5*f/(2*y2-y2/(3.5*f)*2.5*f)*y2-0.05], [0, y2-y2/(3.5*f)*2.5*f], 'c', linewidth=2, zorder=5)
    ax.arrow(-2.5*f+2.5*f/(2*y2-y2/(3.5*f)*2.5*f)*y2-0.05, y2-y2/(3.5*f)*2.5*f-0.1, 0, 0,  head_width=0.07, head_length=0.1, fc='c', ec='none')

    # # Ray 3
    ax.plot([-2.5*f, -2.5*f+2.5*f/(2*y2-y2/(3.5*f)*2.5*f)*y2-0.05],[y2, y2-y2/(3.5*f)*2.5*f], 'r--', linewidth=2)
    ax.plot([-2.5*f+2.5*f/(2*y2-y2/(3.5*f)*2.5*f)*y2-0.05, 2.5*f],[y2-y2/(3.5*f)*2.5*f,-y2], 'r', linewidth=2)
    ax.arrow(f, -y2/2.5, 0.0025, -0.001,  head_width=0.04, head_length=0.06, fc='w', ec='none', zorder=5)

    # # Ray 4
    ax.plot([-2.5*f+2.5*f/(2*y2-y2/(3.5*f)*2.5*f)*y2-0.05, d/2],[y2-y2/(3.5*f)*2.5*f+d/5,y2], 'r', linewidth=2)
    ax.plot([-2.5*f+2.5*f/(2*y2-y2/(3.5*f)*2.5*f)*y2-0.05, -f],[y2-y2/(3.5*f)*2.5*f+d/5,0], 'r--', linewidth=2)

    # f distance
    ax.plot([0, 0], [0, -0.1], 'w-', linewidth=1)
    ax.plot([0, f], [-0.05, -0.05], 'w-', linewidth=1)
    ax.plot([f, f], [0, -0.1], 'w-', linewidth=1)
    ax.plot([0, -f], [-0.05, -0.05], 'w-', linewidth=1)
    ax.plot([-f, -f], [0, -0.1], 'w-', linewidth=1)
    ax.text(f/2, -0.15, r'$f$', horizontalalignment='center', verticalalignment='center', fontsize=fontsize, color='w')
    ax.text(-f/2, -0.15, r'$f$', horizontalalignment='center', verticalalignment='center', fontsize=fontsize, color='w')
    ax.arrow(0.06, -0.05, -0.001, 0,  head_width=0.04, head_length=0.06, fc='w', ec='none')
    ax.arrow(f-0.06, -0.05, 0.001, 0,  head_width=0.04, head_length=0.06, fc='w', ec='none')
    ax.arrow(-0.06, -0.05, 0.001, 0,  head_width=0.04, head_length=0.06, fc='w', ec='none')
    ax.arrow(-f+0.06, -0.05, -0.001, 0,  head_width=0.04, head_length=0.06, fc='w', ec='none')

    if annotation:
        # s distance
        ax.plot([0, 0], [y2-(2.5*f-d/2)/1.5, y2-(2.5*f-d/2)/1.5-0.1], 'w-', linewidth=1)
        ax.plot([0, (f-d/2)/y2*(2.5*f-d/2)/1.5+d/2], [y2-(2.5*f-d/2)/1.5-0.05, y2-(2.5*f-d/2)/1.5-0.05], 'w-', linewidth=1)
        ax.plot([(f-d/2)/y2*(2.5*f-d/2)/1.5+d/2, (f-d/2)/y2*(2.5*f-d/2)/1.5+d/2], [y2-(2.5*f-d/2)/1.5, y2-(2.5*f-d/2)/1.5-0.1], 'w-', linewidth=1)
        ax.plot([0, -2.5*f], [y2-(2.5*f-d/2)/1.5-0.05, y2-(2.5*f-d/2)/1.5-0.05], 'w-', linewidth=1)
        ax.plot([-2.5*f, -2.5*f], [y2-(2.5*f-d/2)/1.5, y2-(2.5*f-d/2)/1.5-0.1], 'w-', linewidth=1)
        ax.text(((f-d/2)/y2*(2.5*f-d/2)/1.5+d/2)/2, y2-(2.5*f-d/2)/1.5-0.15, r'$s_i$', horizontalalignment='center', verticalalignment='center', fontsize=fontsize, color='w')
        ax.text(-2.5*f/2, y2-(2.5*f-d/2)/1.5-0.15, r'$s_o$', horizontalalignment='center', verticalalignment='center', fontsize=fontsize, color='w')
        ax.arrow(0.06, y2-(2.5*f-d/2)/1.5-0.05, -0.001, 0,  head_width=0.04, head_length=0.06, fc='w', ec='none')
        ax.arrow((f-d/2)/y2*(2.5*f-d/2)/1.5+d/2-0.06, y2-(2.5*f-d/2)/1.5-0.05, 0.001, 0,  head_width=0.04, head_length=0.06, fc='w', ec='none')
        ax.arrow(-0.06, y2-(2.5*f-d/2)/1.5-0.05, 0.001, 0,  head_width=0.04, head_length=0.06, fc='w', ec='none')
        ax.arrow(-2.5*f+0.06, y2-(2.5*f-d/2)/1.5-0.05, -0.001, 0,  head_width=0.04, head_length=0.06, fc='w', ec='none')

    # Lens
    lens = Ellipse((0,0), d, 1.2, edgecolor='none', facecolor='skyblue', zorder=-5)
    ax.add_patch(lens)

    # Set limits and aspect ratio
    ax.set_xlim(-1.2, 1.2)
    ax.set_ylim(-0.7, 0.7)
    ax.set_aspect('equal')
    # ax.grid(True)
    ax.axis('off')

    if title:
        ax.set_title(title, loc=title_loc, fontsize=fontsize)

def concave_lens_imaging_virtual(ax, title=None, title_loc='center', fontsize=12, annotation=True):
    # Optical axis
    ax.plot([-1.2, 1.2], [0, 0], 'w', linewidth=2)

    # Distances
    d = 0.05
    f = 0.4

    # y1 and arrows
    y1 = 0.4
    ax.plot([-2.5*f, -2.5*f], [0, y1], 'c', linewidth=2)
    ax.arrow(-2.5*f, y1-0.1, 0, 0,  head_width=0.07, head_length=0.1, fc='c', ec='none')

    # # Angles
    # theta_1 = 0
    # theta_2 = np.arctan2(y1, f-d/2)

    # Ray 1 (horizontal to lens)
    ax.plot([-2.5*f, -d/2], [y1, y1], 'r', linewidth=2)

    # Ray 2 (horizontal after lens)
    ax.plot([-2.5*f, 0], [y1, y1-y1/(3.5*f)*2.5*f], 'r', linewidth=2)
    ax.plot([0, f], [y1-y1/(3.5*f)*2.5*f, 0], 'r--', linewidth=2)
    ax.plot([0, 2.5*f], [y1-y1/(3.5*f)*2.5*f, y1-y1/(3.5*f)*2.5*f], 'r', linewidth=2)
    ax.plot([-2.5*f+2.5*f/(2*y1-y1/(3.5*f)*2.5*f)*y1-0.05, 0], [y1-y1/(3.5*f)*2.5*f, y1-y1/(3.5*f)*2.5*f], 'r--', linewidth=2)
    ax.arrow(f, y1-y1/(3.5*f)*2.5*f, 0.001, 0,  head_width=0.04, head_length=0.06, fc='w', ec='none', zorder=5)

    # y2 and arrows
    ax.plot([-2.5*f+2.5*f/(2*y1-y1/(3.5*f)*2.5*f)*y1-0.05, -2.5*f+2.5*f/(2*y1-y1/(3.5*f)*2.5*f)*y1-0.05], [0, y1-y1/(3.5*f)*2.5*f], 'lime', linewidth=2, zorder=5)
    ax.arrow(-2.5*f+2.5*f/(2*y1-y1/(3.5*f)*2.5*f)*y1-0.05, y1-y1/(3.5*f)*2.5*f-0.1, 0, 0,  head_width=0.07, head_length=0.1, fc='lime', ec='none')

    # # Ray 3 (passing through lens center)
    ax.plot([-2.5*f, 0],[y1, 0], 'r', linewidth=2)
    ax.plot([0, 2.5*f],[0,-y1], 'r', linewidth=2)
    ax.arrow(f, -y1/2.5, 0.0025, -0.001,  head_width=0.04, head_length=0.06, fc='w', ec='none', zorder=5)

    # # Ray 4 (virtual ray from front focal point)
    ax.plot([-2.5*f+2.5*f/(2*y1-y1/(3.5*f)*2.5*f)*y1-0.05, d/2],[y1-y1/(3.5*f)*2.5*f+d/5,y1], 'r--', linewidth=2)
    ax.plot([-2.5*f+2.5*f/(2*y1-y1/(3.5*f)*2.5*f)*y1-0.05, -f],[y1-y1/(3.5*f)*2.5*f+d/5,0], 'r--', linewidth=2)
    ax.plot([d/2, f],[y1,2*y1], 'r', linewidth=2)
    ax.arrow(f/2+d/4, y1+y1/2, 0.001*f/2, 0.001*y1/2,  head_width=0.04, head_length=0.06, fc='w', ec='none', zorder=5)

    # f distance
    ax.plot([0, 0], [0, -0.1], 'w-', linewidth=1)
    ax.plot([0, f], [-0.05, -0.05], 'w-', linewidth=1)
    ax.plot([f, f], [0, -0.1], 'w-', linewidth=1)
    ax.plot([0, -f], [-0.05, -0.05], 'w-', linewidth=1)
    ax.plot([-f, -f], [0, -0.1], 'w-', linewidth=1)
    ax.text(f/2, -0.15, r'$f$', horizontalalignment='center', verticalalignment='center', fontsize=fontsize, color='w')
    ax.text(-f/2, -0.15, r'$f$', horizontalalignment='center', verticalalignment='center', fontsize=fontsize, color='w')
    ax.arrow(0.06, -0.05, -0.001, 0,  head_width=0.04, head_length=0.06, fc='w', ec='none')
    ax.arrow(f-0.06, -0.05, 0.001, 0,  head_width=0.04, head_length=0.06, fc='w', ec='none')
    ax.arrow(-0.06, -0.05, 0.001, 0,  head_width=0.04, head_length=0.06, fc='w', ec='none')
    ax.arrow(-f+0.06, -0.05, -0.001, 0,  head_width=0.04, head_length=0.06, fc='w', ec='none')

    if annotation:
        # s distance
        ax.plot([0, 0], [y1-(2.5*f-d/2)/1.5, y1-(2.5*f-d/2)/1.5-0.1], 'w-', linewidth=1)
        ax.plot([0, (f-d/2)/y1*(2.5*f-d/2)/1.5+d/2], [y1-(2.5*f-d/2)/1.5-0.05, y1-(2.5*f-d/2)/1.5-0.05], 'w-', linewidth=1)
        ax.plot([(f-d/2)/y1*(2.5*f-d/2)/1.5+d/2, (f-d/2)/y1*(2.5*f-d/2)/1.5+d/2], [y1-(2.5*f-d/2)/1.5, y1-(2.5*f-d/2)/1.5-0.1], 'w-', linewidth=1)
        ax.plot([0, -2.5*f], [y1-(2.5*f-d/2)/1.5-0.05, y1-(2.5*f-d/2)/1.5-0.05], 'w-', linewidth=1)
        ax.plot([-2.5*f, -2.5*f], [y1-(2.5*f-d/2)/1.5, y1-(2.5*f-d/2)/1.5-0.1], 'w-', linewidth=1)
        ax.text(((f-d/2)/y1*(2.5*f-d/2)/1.5+d/2)/2, y1-(2.5*f-d/2)/1.5-0.15, r'$s_i$', horizontalalignment='center', verticalalignment='center', fontsize=fontsize, color='w')
        ax.text(-2.5*f/2, y1-(2.5*f-d/2)/1.5-0.15, r'$s_o$', horizontalalignment='center', verticalalignment='center', fontsize=fontsize, color='w')
        ax.arrow(0.06, y1-(2.5*f-d/2)/1.5-0.05, -0.001, 0,  head_width=0.04, head_length=0.06, fc='w', ec='none')
        ax.arrow((f-d/2)/y1*(2.5*f-d/2)/1.5+d/2-0.06, y1-(2.5*f-d/2)/1.5-0.05, 0.001, 0,  head_width=0.04, head_length=0.06, fc='w', ec='none')
        ax.arrow(-0.06, y1-(2.5*f-d/2)/1.5-0.05, 0.001, 0,  head_width=0.04, head_length=0.06, fc='w', ec='none')
        ax.arrow(-2.5*f+0.06, y1-(2.5*f-d/2)/1.5-0.05, -0.001, 0,  head_width=0.04, head_length=0.06, fc='w', ec='none')

    # # Lens
    # lens = Ellipse((0,0), d, 1.2, edgecolor='none', facecolor='skyblue', zorder=-5)
    # ax.add_patch(lens)

    # Set limits and aspect ratio
    ax.set_xlim(-1.2, 1.2)
    ax.set_ylim(-0.7, 0.7)
    ax.set_aspect('equal')
    # ax.grid(True)
    ax.axis('off')

    r=5
    y_concave = np.linspace(-0.6, 0.6, 50)
    x1_concave = -np.sqrt((r-d/2)**2-y_concave**2)+r
    x2_concave = -x1_concave
    plt.fill_betweenx(y_concave, x1_concave, x2_concave, color='skyblue')

    if title:
        ax.set_title(title, loc=title_loc, fontsize=fontsize)

fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(12, 5))

convex_lens_imaging_real(axes[0], title='Convex Lens, Real Image', annotation=False)
convex_lens_imaging_virtual(axes[1], title='Convex Lens, Virtual Image', annotation=False)
concave_lens_imaging_virtual(axes[2], title='Concave Lens, Virtual Image', annotation=False)

plt.tight_layout()
plt.show()
```
