#### Hilbert transform

Hilbert transform of a function $v(x)$ is defined as

$$Hv(x)=\frac{1}{\pi x}*v(x)=\frac{1}{\pi}\int \frac{v(y)}{x-y}dy=-\frac{1}{\pi}\int \frac{v(y)}{y-x}dy$$

Recall Fourier transform of $1/x$ to be

$$F\left(\frac{1}{x}\right)=-\pi i \,\text{sgn}\,s$$

Thus

$$F\left(-\frac{1}{\pi x}\right)=i\, \text{sgn}\,s$$

and Fourier transform of Hilbert transform is

$$F(Hv)(s)=F\left(\frac{1}{\pi x}*v(x)\right)=-i \,(\text{sgn} \,s) V(s)$$

If we apply Hilbert transform twice to a function, and take Fourier transform, we have

$$F(H(Hv))(s)=F\left(\frac{1}{\pi x}*\frac{1}{\pi x}*v(x)\right)=i \,(\text{sgn} \,s)i \,(\text{sgn} \,s) V(s)=-V(s)$$

Take inverse Fourier transform, we have

$$H(Hv(x))=-v(x)$$

This means that inverse Hilbert transform is just its negative

$$H^{-1}v(x)=-Hv(x)=-\frac{1}{\pi x}*v(x)$$

#### Analytic signals

The use of Hilbert transform allows to complete a real signal to an `analytic signal` â€” a complex-valued form that involves only positive frequencies

This representation can simplify results and methods in some cases, especially in applications to communications systems

Let $u$ be the unit step function, then `by construction`, the following

$$z(t)=F^{-1}(2u(s)V(s))$$

has no negative frequencies in its spectrum

We can see how it is related to the original signal $v(t)$

$$
\begin{align*}
z(t)&=F^{-1}(2u(s)V(s)) \\
&=F^{-1}((1+\text{sgn }s)V(s)) \\
&=F^{-1}(V(s)+(\text{sgn }s) \,V(s)) \\
&=F^{-1}(V(s))+F^{-1}((\text{sgn }s) \,V(s)) \\
&=v(t) + i (Hv)(t)
\end{align*}
$$

$z(t)$ is called the `analytic signal` associated with $v(t)$, and $v(t) = \text{Re}\,z(t)$

Briefly, to get the spectrum of $z(t)$, cut off the negative frequencies of $v(t)$ and double the amplitudes of the positive frequencies

#### Narrowband signals

Narrowband signal is often in the form of

$$v(t)=a(t)\cos (2\pi f_0 t+\phi(t))$$

where the amplitude $a(t)$ and phase $\phi(t)$ are `varying slowly`, around zero in spectrum, far below $f_0$

We can write $v(t)$ in exponential form

$$v(t)=\frac{1}{2}a(t)e^{i\phi(t)}e^{2\pi i f_0 t}+\frac{1}{2}a(t)e^{-i\phi(t)}e^{-2\pi i f_0 t}$$

Remember that we get analytic signal corresponding to $v(t)$ by cutting off the negative frequencies and doubling the amplitudes of the positive frequencies

So to the extent that the spectral islands of $v(t)$ are really separated, and the amplitude and phase are slowly varying functions, we are justified in saying that, approximately,

$$z(t) = a(t)e^{i\phi(t)}e^{2\pi i f_0 t}$$

where $a(t)$ is the `envelope` and $a(t)=|z(t)|$

`phase` is $\psi(t)=2\pi f_0t + \phi(t)$

and `instantaneous frequency` is

$$f_{\text{inst}}=\frac{1}{2\pi} \frac{d \psi}{dt}=f_0+\frac{1}{2\pi}\frac{d\phi}{dt}$$

#### Defects in machinery

We see an application of Hilbert transform for detects in machinery

**A** Periodic impacts induced by defects excite a high-frequency resonance in machine structure

Let impacts arrive every $T$ seconds ($f_m= 1/T$)

$$p(t)=\sum_{n\in \mathbb{Z}}\delta (t-nT)$$

Let structural impulse reponse at excited resonance $f_c \gg f_m$

$$g(t)=\psi(t) \cos (2\pi f_c t), \psi(t)=e^{-\alpha t}u(t)$$

The raw response for this is a periodized signal through convolution

$$r(t)=(g*p)(t) = \sum_{n}g(t-nT)$$

**B** Separate `carrier` and `envelope`

Write out more explicitly

$$
\begin{align*}
r(t)&=\sum_n \psi(t-nT) \cos\left[2\pi f_c (t-nT)\right] \\
&=a_c(t)\cos(2\pi f_c t)+ a_s(t) \sin (2\pi f_c t)
\end{align*}$$

where

$$\begin{align*}
a_c(t)&=\sum_n \psi(t-nT) \cos (2\pi f_c nT)\\
a_s(t)&=\sum_n \psi(t-nT) \sin (2\pi f_c nT)
\end{align*}$$

We see $a_c(t), a_s(t)$ are $T$-periodic

For a `narrow band` around $f_c$, we can bundle them into a a single slowly varying `envelope` $a(t)$ and phase $\phi(t)$

$$\begin{align*}x(t) &\approx a(t)\cos(2\pi f_c t + \phi(t))\\
a(t) &= \sqrt{a_c(t)^2+a_s(t)^2} \\
(t+T)&=a(t) \\
\phi(t) &= \tan^{-1}(a_s(t)/a_c(t))\end{align*}$$

If we only have a single resonance mode at $f_c$, then $x(t)=r(t)$, no bandpass needed

Real measurements however, contain multiple modes, therefore

$$\begin{align*}r(t)&=\sum_m r_m(t) + \text{other stuff}\\
r_m(t)&\approx a_m(t) \cos(2\pi f_{c,m}t +\phi_m(t))
\end{align*}$$

To get a clean picture for mode $m$, we use bandpass filter to form

$$x(t)=\text{BP}_{f_{c,m}}[r(t)]\approx r_m(t)$$

**C** Envelope has a Fourier series at fault rate

With $a(t)$ being $T$-periodic, we write it using Fourier series

$$a(t)=\sum_{k\in \mathbb{Z}}a_k e^{2\pi i f_m k t}, a_k = \frac{1}{T}\int_{0}^Ta(t) e^{-2 \pi i f_m k t}$$

**D** `Sidebands` in Fourier domain

Use complex form (drop phase term, which won't affect sideband centered around $f_c$ since it would go to complex coefficients)

$$
\begin{align*}
x(t)&=\frac{1}{2}a(t) e^{i\phi(t)} e^{2 \pi i f_c t}+\frac{1}{2}a(t) e^{-i\phi(t)}e^{-2 \pi i f_c t} \\
& \text{drop phase term for clarity} \\
&=\frac{1}{2}\sum_{k}a_k e^{2\pi i(f_c+kf_m)t}+\frac{1}{2}\sum_{k}a_k e^{2\pi i(-f_c+kf_m)t}
\end{align*}$$

Given $F(e^{2 \pi i a t})=\delta(f-a)$, we have the Fourier transform of $x(t)$ as

$$X(f)=\frac{1}{2}\sum_k a_k \delta\left(f-(f_c+kf_m)\right)+\frac{1}{2}\sum_k a_k \delta\left(f-(-f_c+kf_m)\right)$$

That's the `sideband` comb, lines at $f_c\pm kf_m$

**E** Hilbert transform and analytic signal

The sidebands are in the Fourier spectrum, but Hilbert transform is what lets us demodulate cleanly and pull out the envelope and instantaneous phase/frequency, which are often much easier to analyze and more robust in practice

We know already the analytic signal is

$$z(t)=x(t)+iHx(t)$$

Since $a(t), \phi(t)$ vary slowly (usually the case), then

$$z(t)\approx a(t) e^{i(2\pi f_c t+\phi(t))}$$

and $|z(t)|=a(t)$, which is the envelope

We do Fourier transform of $a(t)$ (after gettting rid of mean) to get lines at $kf_m$

```python
import numpy as np
from scipy.signal import hilbert, butter, filtfilt, spectrogram, windows, sosfiltfilt
import matplotlib.pyplot as plt
```

```python
def bandpass_sos(x, fs, lowcut, highcut, order=4):
    x = np.asarray(x, float)
    x = x - x.mean()
    sos = butter(order, [lowcut, highcut], btype='band', fs=fs, output='sos')
    y  = sosfiltfilt(sos, x)
    return y

def hilbert_envelope(x):
    z = hilbert(x)
    a = np.abs(z)
    return a

def envelope_spectrum(x, fs, lowcut, highcut, order=4, use_hann=True, trim_edges_frac=0.0):
    y = bandpass_sos(x, fs, lowcut, highcut, order)

    a = hilbert_envelope(y)
    a = a - a.mean()

    if trim_edges_frac > 0:
        n = len(a)
        k = int(n * trim_edges_frac / 2)
        a = a[k:n-k]

    if use_hann:
        w = windows.hann(len(a))
        cg = 0.5
        a = (a * w) / cg

    N = len(a)
    F = np.fft.rfftfreq(N, d=1/fs)
    A = (2.0 / N) * np.abs(np.fft.rfft(a))
    return F, A, a
```

```python
fs = 200000.0
dt = 1/fs
T_dur = 0.20
fc = 5000.0
fm = 120.0
T = 1.0 / fm
alpha = 2*np.pi*400
trim_edges_frac = 0.05

t = np.arange(0, T_dur, dt)
N = t.size

n_impacts = int(np.floor(T_dur * fm)) + 1
impact_times = np.arange(n_impacts) * T

r = np.zeros_like(t)
a_c = np.zeros_like(t)
a_s = np.zeros_like(t)

for tn in impact_times:
    tau = t - tn
    u = (tau >= 0).astype(float)
    psi = np.exp(-alpha * tau) * u
    r += psi * np.cos(2*np.pi*fc * (t - tn))
    a_c += psi * np.cos(2*np.pi*fc * tn)
    a_s += psi * np.sin(2*np.pi*fc * tn)

a_true = np.sqrt(a_c**2 + a_s**2)
phi_true = np.arctan2(a_s, a_c)
r_recon = a_true * np.cos(2*np.pi*fc*t - phi_true)

lowcut = fc - 800.0
highcut = fc + 800.0
F_env, A_env, a_bp_hilb = envelope_spectrum(
    r, fs, lowcut=lowcut, highcut=highcut, order=4, use_hann=True, trim_edges_frac=trim_edges_frac
)

# 1) Time overlay
t_zoom_max = 0.040
idx_zoom = t <= t_zoom_max

plt.figure(figsize=(10, 4))
plt.plot(t[idx_zoom]*1e3, r[idx_zoom], label="r(t)")
plt.plot(t[idx_zoom]*1e3, a_true[idx_zoom], linestyle="--", label="+ a_true(t)")
plt.plot(t[idx_zoom]*1e3, -a_true[idx_zoom], linestyle="--", label="- a_true(t)")
trim = int(trim_edges_frac * len(t) / 2)
t_env = t[trim:len(t)-trim]
idx_zoom_env = t_env <= t_zoom_max
plt.plot(t_env[idx_zoom_env]*1e3, a_bp_hilb[idx_zoom_env], label="BP+Hilbert envelope")
plt.xlabel("Time (ms)")
plt.ylabel("Amplitude (linear)")
plt.title("Signal with exact envelope and BP+Hilbert envelope (zoom)")
plt.legend()
plt.tight_layout()
plt.show()

# 2) Spectrum near carrier
r0 = r - r.mean()
R = np.fft.rfft(r0) / N
freq = np.fft.rfftfreq(N, d=dt)
A_raw = (2.0 / N) * np.abs(R)  # single-sided linear amplitude
band = (freq > (fc - 1000)) & (freq < (fc + 1000))

plt.figure(figsize=(10, 4))
plt.plot(freq[band], A_raw[band])
plt.xlabel("Frequency (Hz)")
plt.ylabel("Amplitude (linear)")
plt.title("Sidebands near the carrier (linear amplitude)")
plt.tight_layout()
plt.show()

# 3) Envelope spectrum (baseband)
mask_env = (F_env >= 0) & (F_env <= 1000)
plt.figure(figsize=(10, 4))
plt.plot(F_env[mask_env], A_env[mask_env])
plt.xlabel("Frequency (Hz)")
plt.ylabel("Amplitude (linear)")
plt.title("Envelope spectrum after BP + Hilbert (linear amplitude)")
plt.tight_layout()
plt.show()

# 4) Peak list
def top_peaks_linear(f, A, fmax=1000.0, k=8, min_sep=20.0):
    mask = (f >= 0) & (f <= fmax)
    f = f[mask]; A = A[mask]
    idx_sorted = np.argsort(A)[::-1]
    chosen = []
    for i in idx_sorted:
        fi = f[i]
        if all(abs(fi - f[j]) >= min_sep for j in chosen):
            chosen.append(i)
        if len(chosen) >= k:
            break
    return [(float(f[i]), float(A[i])) for i in chosen]

peaks_linear = top_peaks_linear(F_env, A_env, fmax=1000.0, k=10, min_sep=20.0)
peaks_linear
```

#### Generalized demodulation

For a signal $x(t)$, its generalized Fourier transform is

$$X_G(f)=\int x(t) e^{-2\pi i [ft+p_0(t)]}dt$$

where $p_0(t)$ denotes cumulative phase of signal, or a `phase function`

The corresponding generalized inverse Fourier transform is

$$\begin{align*}
x(t)&=\frac{1}{2\pi}\int X_G(f)e^{2\pi i [ft+p_0(t)]} df \\
&=\frac{1}{2\pi}e^{2\pi i p_0(t)} \int X_G(f) e^{2\pi i ft}df
\end{align*}$$

##### Time-varying frequency to constant

Now, if what we want to map a time-varying frequency $f(t)$ to a constant frequency $f_0$ in $X_G(f)$, by design, what we want is

$$X_G(f)=2\pi \delta (f-f_0)$$

Plug this into generalized inverse Fourier transform, we have

$$x(t) = \frac{1}{2\pi}e^{2\pi i p_0(t)}\cdot 2 \pi e^{2\pi i f_0 t}=e^{2\pi i[f_0t+p_0(t)]}$$

and `instantaneous frequency` is

$$f(t) =f_0 +\frac{dp_0(t)}{dt}$$

and the phase function is computed as

$$\frac{dp_0(t)}{dt}=f(t) - f_0$$

$$p_0(t)=\int_0^t [f(t)-f_0] dt$$

##### Demodulation

Therefore, for a real signal x(t) with time-varying instantaneous frequency $f(t)$, by denoting

$$y(t) = x(t) e^{-2 \pi i p_0(t)}$$

$y(t)$ is the `demodulated signal` with only a constant component at $f_0$

The only thing we need to do is to approximate $f(t)$

##### Polynomial phase funtion

One option is to approximate $f(t)$ via polynomial

$$f(t) = \sum_{n=1}^{k-1} \alpha_n t^n +\alpha_0$$

and

$$p_0(t)=\sum_{n=1}^{k-1} \frac{\alpha_n}{n+1}t^{n+1} +(\alpha_0-f_0)t$$

##### Real signal

The above derivation assumes $x(t)$ is complex, however, for real (periodic) signal $x(t)$ we can measure, it can always be written as

$$x(t)=\frac{1}{2}e^{2 \pi i [f_0t +p_0(t)]}+\frac{1}{2}e^{-2 \pi i [f_0t +p_0(t)]}$$

Therefore, directly multiplying with $e^{-2\pi i p_0(t)}$ can cause problem, as

$$x(t)e^{-2\pi i p_0(t)} = \frac{1}{2}e^{2 \pi i f_0t}+\frac{1}{2}e^{-2 \pi i [f_0t +2p_0(t)]}$$

where only the first term is what we want

As a result, we want to remove the negative frequency part of $x(t)$ first, one way to do so is to simply use Hilbert transform before multiplication with $e^{-2\pi i p_0(t)}$ (of course, this doubles energy, but usually scale of energy is less of a concern for spectrum analysis)

##### Scaling

Finally, if we want to demodulate $cf(t)$, then we need use $cf(t)$ to get scaled version that is $cp_0(t)$

```python
fs = 2000 # Sampling frequency
T = 3 # Duration (s)
t = np.linspace(0, T, int(T*fs))
c = 2

# True instantaneous frequency
f_start, f_end = 50, 200
f_t = np.linspace(f_start, f_end, len(t)) # true f(t)
cf_t = c * f_t # fault frequency proportional to f_t

# Generate signal with time-varying phase
phi_t = np.cumsum(f_t) / fs
cphi_t = np.cumsum(cf_t) / fs
x = np.exp(1j * 2 * np.pi * phi_t) + 0.6 * np.exp(1j * 2 * np.pi * cphi_t) # analytic signal
x_real = np.real(x) + 0.3 * np.random.randn(len(t))  # noisy real signal

measured_freq = f_t + 5 * np.random.randn(len(t)) # noisy measurement

# Fit a polynomial to approximate instantaneous frequency
coeffs = np.polyfit(t, measured_freq, 2)
fitted_freq = np.polyval(coeffs, t)

# p_0(t) = int [f(t) - f_0] dt
f_0 = 100.0 # base (reference) frequency
p_0_t = np.cumsum(fitted_freq - f_0) / fs # numerical integration (dt = 1/fs)

# demodulate for C=1 and C=2
y1 = hilbert(x_real) * np.exp(-1j * 2 * np.pi * p_0_t) # flatten f_r to f0
y2 = hilbert(x_real) * np.exp(-1j * 2 * np.pi * c * p_0_t) # flatten cf_r to cf0

# spectrograms
f, t_spec, Sx = spectrogram(x_real, fs, nperseg=256, noverlap=200)
_, _, Sy1 = spectrogram(np.real(y1), fs, nperseg=256, noverlap=200)
_, _, Sy2 = spectrogram(np.real(y2), fs, nperseg=256, noverlap=200)

plt.figure(figsize=(12, 10))
plt.subplot(3,1,1)
plt.pcolormesh(t_spec, f, 10*np.log10(Sx+1e-8), shading='gouraud')
plt.title('Original signal: f_t and cf_t (rising)')
plt.ylabel('Frequency [Hz]')

plt.subplot(3,1,2)
plt.pcolormesh(t_spec, f, 10*np.log10(Sy1+1e-8), shading='gouraud')
plt.title('After GD, demodulated to f_0')
plt.ylabel('Frequency [Hz]')

plt.subplot(3,1,3)
plt.pcolormesh(t_spec, f, 10*np.log10(Sy2+1e-8), shading='gouraud')
plt.title('After GD, demodulated to cf_0')
plt.xlabel('Time [s]')
plt.ylabel('Frequency [Hz]')

plt.tight_layout()
plt.show()
```

```python
X_fft = np.fft.fft(x_real)
Y1_fft = np.fft.fft(np.real(y1))
Y2_fft = np.fft.fft(np.real(y2))
freqs = np.fft.fftfreq(len(t), d=1/fs)

# Use only positive frequencies
pos_mask = freqs >= 0
freqs = freqs[pos_mask]
X_mag = np.abs(X_fft[pos_mask])
Y1_mag = np.abs(Y1_fft[pos_mask])
Y2_mag = np.abs(Y2_fft[pos_mask])

plt.figure(figsize=(10, 5))
plt.plot(freqs, X_mag / np.max(X_mag), label='Original signal')
plt.plot(freqs, Y1_mag / np.max(Y1_mag), label='After GD (demodulated to f_0)')
plt.plot(freqs, Y2_mag / np.max(Y2_mag), label='After GD (demodulated to cf_0)')
# plt.axvline(f_0, color='r', linestyle='--', label='Base freq f0', alpha=0.5)
plt.xlabel('Frequency [Hz]')
plt.ylabel('Normalized amplitude')
plt.title('FFT before and after Generalized Demodulation')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

##### Demodulation of enveloped signal

```python
fs = 20000
T = 5.0
t = np.linspace(0, T, int(fs*T))

# Resonance frequency
f_c = 2000

f_r = np.linspace(30, 60, len(t))

# Fault constant (BPFO order)
C_bpfo = 3.2
f_bpfo_t = C_bpfo * f_r # time-varying BPFO

# Envelope modulation
phi_m = 2 * np.pi * np.cumsum(f_bpfo_t) / fs
a_t = 1 + 0.4 * np.cos(phi_m)

# Raw vibration signal (modulated resonance)
x = a_t * np.cos(2 * np.pi * f_c * t)
x += 2 * np.random.randn(len(t))

def bandpass(sig, fs, low, high, order=4):
    b, a = butter(order, [low/(fs/2), high/(fs/2)], btype='band')
    return filtfilt(b, a, sig)

x_bp = bandpass(x, fs, 1500, 2500)

# Envelope of band-passed signal
a_env = np.abs(hilbert(x_bp))
a_env_no_dc = a_env - np.mean(a_env)

# Make envelope analytic and demodulate
a_env_an = hilbert(a_env)
f0 = 40.0 # base frequency
p0_bpfo = np.cumsum(f_bpfo_t - C_bpfo*f0) / fs # phase function

a_env_gd = a_env_an * np.exp(-1j * 2*np.pi * p0_bpfo) # demodulation

# Remove DC and apply window for FFT
a_env_gd_no_dc = np.real(a_env_gd) - np.mean(np.real(a_env_gd))
window = np.hanning(len(a_env_gd_no_dc))
sig = a_env_gd_no_dc * window

def fft_mag(sig, fs):
    N = len(sig)
    freqs = np.fft.rfftfreq(N, 1/fs)
    mag = np.abs(np.fft.rfft(sig))
    mag /= np.max(mag)
    return freqs, mag

freqs_raw, mag_raw = fft_mag(a_env_no_dc, fs)
freqs_gd, mag_gd = fft_mag(a_env_gd_no_dc, fs)

plt.figure(figsize=(10,5))
plt.plot(freqs_raw, mag_raw, label='Raw envelope (no GD)', alpha=0.6)
plt.plot(freqs_gd, mag_gd, label=f'GD on envelope: BPFO = {C_bpfo*f0:.1f} Hz', linewidth=1.5)
plt.xlim(0, 250)
plt.xlabel('Frequency [Hz]')
plt.ylabel('Normalized magnitude')
plt.title('Envelope FFT: Before and After Generalized Demodulation')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
