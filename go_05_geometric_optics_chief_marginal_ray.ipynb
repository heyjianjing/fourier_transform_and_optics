### Chief ray and marginal ray

```python
import sympy as smp
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Arc, Rectangle, Ellipse, Circle
import numpy as np

np.random.seed(42)
np.set_printoptions(formatter={'float': '{: 0.4f}'.format})

plt.style.use('dark_background')

# color: https://matplotlib.org/stable/gallery/color/named_colors.html
```

```python
def chief_marginal_ray(ax, title=None, title_loc='center', fontsize=12):
    # Distances
    d = 0.05
    f = 0.4

    # lens 1
    lens_1 = Ellipse((-2*f,0), 0.1, 0.7, edgecolor='none', facecolor='skyblue', zorder=-5)
    ax.add_patch(lens_1)

    # lens 2
    eyepiece = Ellipse((2*f,0), 0.1, 0.7, edgecolor='none', facecolor='skyblue', zorder=-5)
    ax.add_patch(eyepiece)

    # Ray 1 (start from center)
    ax.plot([-4*f, 5*f], [0, 0], 'r', linewidth=2)
    ax.plot([-4*f, -2*f],[0, f/4], 'r', linewidth=2, alpha=0.2)
    ax.plot([-4*f, -2*f],[0, -f/4], 'r', linewidth=2, alpha=0.2)
    ax.plot([-2*f, 2*f],[f/4, f/4], 'r', linewidth=2, alpha=0.2)
    ax.plot([-2*f, 2*f],[-f/4, -f/4], 'r', linewidth=2, alpha=0.2)
    ax.plot([2*f, 5*f],[f/4, -f/8], 'r', linewidth=2, alpha=0.2)
    ax.plot([2*f, 5*f],[-f/4, f/8], 'r', linewidth=2, alpha=0.2)
    ax.arrow(-3*f, 0, 0.001, 0,  head_width=0.04, head_length=0.06, fc='r', ec='none', zorder=5)
    source_red = Circle((-4*f,0), 0.02, edgecolor='none', facecolor='red', zorder=-6)
    ax.add_patch(source_red)
    focus_red = Circle((4*f,0), 0.02, edgecolor='none', facecolor='red', zorder=-6)
    ax.add_patch(focus_red)

    # Ray 2 (start from top)
    ax.plot([-4*f, -2*f], [f/4, f/4], 'c', linewidth=2)
    ax.plot([-4*f, -2*f],[f/4, f/2], 'c', linewidth=2, alpha=0.2)
    ax.plot([-4*f, -2*f],[f/4, 0], 'c', linewidth=2, alpha=0.2)
    ax.plot([-2*f, 2*f],[f/4, -f/4], 'c', linewidth=2)
    ax.plot([2*f, 5*f],[-f/4, -f/4], 'c', linewidth=2)
    ax.plot([-2*f, 5*f],[f/2, -3*f/8], 'c', linewidth=2, alpha=0.2)
    ax.plot([-2*f, 2*f],[0, -f/2], 'c', linewidth=2, alpha=0.2)
    ax.plot([2*f, 5*f],[-f/2, -f/8], 'c', linewidth=2, alpha=0.2)
    ax.arrow(-3*f, f/4, 0.001, 0,  head_width=0.04, head_length=0.06, fc='c', ec='none', zorder=5)
    source_green = Circle((-4*f,f/4), 0.02, edgecolor='none', facecolor='c', zorder=-6)
    ax.add_patch(source_green)
    focus_green = Circle((4*f,-f/4), 0.02, edgecolor='none', facecolor='c', zorder=-6)
    ax.add_patch(focus_green)

    # Ray 3 (start from bottom)
    ax.plot([-4*f, -2*f], [-f/4, -f/4], 'y', linewidth=2)
    ax.plot([-4*f, -2*f],[-f/4, -f/2], 'y', linewidth=2, alpha=0.2)
    ax.plot([-4*f, -2*f],[-f/4, 0], 'y', linewidth=2, alpha=0.2)
    ax.plot([-2*f, 2*f],[-f/4, f/4], 'y', linewidth=2)
    ax.plot([2*f, 5*f],[f/4, f/4], 'y', linewidth=2)
    ax.plot([-2*f, 5*f],[-f/2, 3*f/8], 'y', linewidth=2, alpha=0.2)
    ax.plot([-2*f, 2*f],[0, f/2], 'y', linewidth=2, alpha=0.2)
    ax.plot([2*f, 5*f],[f/2, f/8], 'y', linewidth=2, alpha=0.2)
    ax.arrow(-3*f, -f/4, 0.001, 0,  head_width=0.04, head_length=0.06, fc='y', ec='none', zorder=5)
    source_blue = Circle((-4*f,-f/4), 0.02, edgecolor='none', facecolor='y', zorder=-6)
    ax.add_patch(source_blue)
    focus_blue = Circle((4*f,f/4), 0.02, edgecolor='none', facecolor='y', zorder=-6)
    ax.add_patch(focus_blue)

    # Aperture
    aperture_top = Rectangle((-0.025, f/4+0.01), 0.05, f/2, linewidth=1, edgecolor='none', facecolor='w', alpha=0.9)
    ax.add_patch(aperture_top)
    aperture_bottom = Rectangle((-0.025, -f/4-0.01), 0.05, -f/2, linewidth=1, edgecolor='none', facecolor='w', alpha=0.9)
    ax.add_patch(aperture_bottom)
    ax.text(0, 3*f/4+0.07, 'aperture', horizontalalignment='center', verticalalignment='center', fontsize=fontsize, color='w')

    # Set limits and aspect ratio
    ax.set_xlim(-1.7, 2)
    ax.set_ylim(-0.5, 0.5)
    ax.set_aspect('equal')
    ax.grid(True)
    ax.axis('off')

    if title:
        ax.set_title(title, loc=title_loc, fontsize=fontsize)
```

```python
fig, ax = plt.subplots(figsize=(10, 3))

chief_marginal_ray(ax, title='Chief and Marginal Rays')

plt.tight_layout()
plt.show()
```

Chief rays go through center of aperture \
Marginal rays go pass edge aperture

### Aperture size and depth of focus

```python
def depth_of_focus(ax, title=None, title_loc='center', fontsize=12):
    # Distances
    d = 0.05
    f = 0.4

    # lens 1
    lens_1 = Ellipse((-2*f,0), 0.1, 0.7, edgecolor='none', facecolor='skyblue', zorder=-5)
    ax.add_patch(lens_1)

    # lens 2
    eyepiece = Ellipse((2*f,0), 0.1, 0.7, edgecolor='none', facecolor='skyblue', zorder=-5)
    ax.add_patch(eyepiece)

    # Ray 1 (start from center)
    ax.plot([-4*f, 5*f], [0, 0], 'r', linewidth=2)
    ax.plot([-4*f, -2*f],[0, f/4], 'r', linewidth=2)
    ax.plot([-4*f, -2*f],[0, -f/4], 'r', linewidth=2)
    ax.plot([-2*f, 2*f],[f/4, f/4], 'r', linewidth=2)
    ax.plot([-2*f, 2*f],[-f/4, -f/4], 'r', linewidth=2)
    ax.plot([2*f, 5*f],[f/4, -f/8], 'r', linewidth=2)
    ax.plot([2*f, 5*f],[-f/4, f/8], 'r', linewidth=2)
    ax.plot([-4*f, -2*f], [0, f/2], 'r', linewidth=2, alpha=0.4)
    ax.plot([-4*f, -2*f], [0, -f/2], 'r', linewidth=2, alpha=0.4)
    ax.plot([-2*f, 2*f],[f/2, f/2], 'r', linewidth=2, alpha=0.4)
    ax.plot([-2*f, 2*f],[-f/2, -f/2], 'r', linewidth=2, alpha=0.4)
    ax.plot([2*f, 5*f],[f/2, -f/4], 'r', linewidth=2, alpha=0.4)
    ax.plot([2*f, 5*f],[-f/2, f/4], 'r', linewidth=2, alpha=0.4)
    ax.arrow(-3*f, 0, 0.001, 0,  head_width=0.04, head_length=0.06, fc='r', ec='none', zorder=5)
    source_red = Circle((-4*f,0), 0.02, edgecolor='none', facecolor='red', zorder=-6)
    ax.add_patch(source_red)
    focus_red = Circle((4*f,0), 0.02, edgecolor='none', facecolor='red', zorder=-6)
    ax.add_patch(focus_red)

    # Aperture
    aperture_top = Rectangle((-0.025, f/4+0.01), 0.05, f/2, linewidth=1, edgecolor='none', facecolor='w', alpha=0.9)
    ax.add_patch(aperture_top)
    aperture_bottom = Rectangle((-0.025, -f/4-0.01), 0.05, -f/2, linewidth=1, edgecolor='none', facecolor='w', alpha=0.9)
    ax.add_patch(aperture_bottom)
    ax.text(0, 3*f/4+0.07, 'aperture', horizontalalignment='center', verticalalignment='center', fontsize=fontsize, color='w')

    # f_t distance
    ax.plot([5.1*f, 5.2*f], [f/4, f/4], 'w-', linewidth=1)
    ax.plot([5.1*f, 5.2*f], [-f/4, -f/4], 'w-', linewidth=1)
    ax.plot([5.15*f, 5.15*f], [f/4, -f/4], 'w-', linewidth=1)
    ax.arrow(5.15*f, f/4-0.06, 0, 0,  head_width=0.04, head_length=0.06, fc='w', ec='none')
    ax.arrow(5.15*f, -f/4+0.06, 0, -0.001,  head_width=0.04, head_length=0.06, fc='w', ec='none')
    ax.text(6*f, 0, 'larger aperture \nsmaller DoF', horizontalalignment='center', verticalalignment='center', fontsize=fontsize, color='w')

    # Set limits and aspect ratio
    ax.set_xlim(-1.7, 2.5)
    ax.set_ylim(-0.5, 0.5)
    ax.set_aspect('equal')
    ax.grid(True)
    ax.axis('off')

    if title:
        ax.set_title(title, loc=title_loc, fontsize=fontsize)
```

```python
fig_2, ax_2 = plt.subplots(figsize=(12, 3))

depth_of_focus(ax_2, title='Depth of Focus')

plt.tight_layout()
plt.show()
```
